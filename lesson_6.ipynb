## Задание 1

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
data = boston.data
```

```python
data.shape
```

```python
target = boston.target
```

```python
feature_names = boston.feature_names
```

```python
X = pd.DataFrame(data, columns = feature_names)
```

```python
X.info
```

```python
y = pd.DataFrame(target, columns = ['Price'])
```

```python
y.info
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({'y_test': y_test['Price'],
                           'y_pred': y_pred.flatten()},
                         columns=['y_test', 'y_pred'])
```

```python
check_test.head(10)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

## Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred1 = model.predict(X_test)
```

```python
check_test1 = pd.DataFrame({'y_test': y_test['Price'],
                           'y_pred1': y_pred1.flatten()},
                         columns=['y_test', 'y_pred1'])
```

```python
check_test1.head(10)
```

```python
r2_score(y_test, y_pred1)
```

### Модель RandomForestRegressor работает лучше чем LinearRegression.
### 0.7112260057484943 < 0.87472606157312

## Задание 4

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('/Загрузки/creditcard.csv')
```

```python
df['Class'].value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
X = df.iloc[:, 0:30]
```

```python
X
```

```python
y = pd.Series(df['Class'])
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.info, X_test.info, y_train.info, y_test.info
```

```python
parameters = [{
    'n_estimators': [10, 15], 
    'max_features': np.arange(3, 5), 
    'max_depth': np.arange(4, 7)
}]
```

```python
gscv = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3
)
```

```python
gscv.fit(X_train, y_train)
```

```python
gscv.best_params_
```

```python
clf = RandomForestClassifier(max_depth=6, max_features=3, n_estimators=15)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred_proba = clf.predict_proba(X_test)[:, 1]
```

```python
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
